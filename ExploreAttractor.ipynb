## Integration methods dependence

```python
" This part studies the integration of the euler method"
AttractorObject = Attractor()
AttractorObject.euler([0.1, 0.0, 0.0])
AttractorObject.evolve(order = 1)
AttractorObject.plotx() 
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

```python
" This part studies the integration of the runge kutta 2 method "
AttractorObject = Attractor()
AttractorObject.rk2([0.1, 0.0, 0.0])
AttractorObject.evolve(order = 2)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

```python
" This part studies the integrathion of the runge kutta 4 method"
AttractorObject = Attractor()
AttractorObject.euler([0.1, 0.0, 0.0])
AttractorObject.evolve(order = 4)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

### The above plots demonstrate how does integration method affects the solution. For, three simulations were run for euler (1st order), RK2 (2nd order) and RK4 (4th order), integrators.  As one see, the plots mainly show the time dependence of x,y, and z coordinates. For, RK integration curves are quite similar to those of RK2 despite some minute differences.  On the other hand, we can see a poor performence of the Euler scheme compared to the RK schemes

#  Effects of the initial values

## For, let's just choose one of the method (Rk4) and conduct our analysis

```python
" See how the initial conditions r0 = [0.1,0.0,.0.] can affect the outrunge kutta 4 method "
AttractorObject = Attractor()
AttractorObject.evolve([0.1, 0.0, 0.0],order = 4)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

```python
" See how the initial conditions r0 = [-5,-5,5] can affect the outrunge kutta 4 method "
AttractorObject = Attractor()
AttractorObject.evolve([-5, -5, 5],order = 4)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

```python
" See how the initial conditions r0 = [7,-7,7] can affect the outrunge kutta 4 method "AttractorObject = Attractor()
AttractorObject.evolve([7, 7, 7],order = 4)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

```python
" See how the initial conditions r0 = [-10,-8,-6] can affect the outrunge kutta 4 method "
AttractorObject = Attractor()
AttractorObject.evolve([-10, -8, -6],order = 4)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

```python
" See how the initial conditions r0 = [8,9,10] can affect the outrunge kutta 4 method "
AttractorObject = Attractor()
AttractorObject.evolve([8, 9, 10],order = 4)
AttractorObject.plotx()
AttractorObject.ploty()
AttractorObject.plotz()
AttractorObject.plot3d()
```

### Figures above show how much initial parameters variation affects. the shape of dynamical curves. It is demonstrated that all initial parameters can be changed in range [-5, 7] without any significant shape break. Parallel to those, one can see differences between the x, y and z plots for initial values, but the 3D are quasi-similar. These differences in the x,y, and z plots can preasumably be due to some random noises.
